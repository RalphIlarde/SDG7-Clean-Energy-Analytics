## Converting Wide Data to Long Format

The World Bank datasets store years as separate columns. For regression analysis, I convert the data into a long format where each row represents a country-year observation.

years = [str(year) for year in range(2000, 2022)]

```python
years = [str(year) for year in range(2000, 2022)]

print(years)
```

```python
renewable_long = renewable.melt(
    id_vars=["Country Name", "Country Code"],
    value_vars=years,
    var_name="Year",
    value_name="Renewable_Energy"
)

gdp_long = gdp.melt(
    id_vars=["Country Name", "Country Code"],
    value_vars=years,
    var_name="Year",
    value_name="GDP_per_Capita"
)

urban_long = urban.melt(
    id_vars=["Country Name", "Country Code"],
    value_vars=years,
    var_name="Year",
    value_name="Urban_Population"
)

electricity_long = electricity.melt(
    id_vars=["Country Name", "Country Code"],
    value_vars=years,
    var_name="Year",
    value_name="Electricity_Access"
)
```

```python
renewable_long["Year"] = renewable_long["Year"].astype(int)
gdp_long["Year"] = gdp_long["Year"].astype(int)
urban_long["Year"] = urban_long["Year"].astype(int)
electricity_long["Year"] = electricity_long["Year"].astype(int)
```

## Removing Aggregate Regions and Income Groups

The World Bank datasets include aggregate records such as "World" and income classifications. Since the objective of this study is to compare countries, these aggregate observations are removed.

```python
exclude_codes = [
    "WLD",
    "HIC",
    "LIC",
    "LMC",
    "UMC",
    "EAS",
    "ECS",
    "LCN",
    "MEA",
    "NAC",
    "SAS",
    "SSF"
]
```

```python
renewable_long = renewable_long[
    ~renewable_long["Country Code"].isin(exclude_codes)
]

gdp_long = gdp_long[
    ~gdp_long["Country Code"].isin(exclude_codes)
]

urban_long = urban_long[
    ~urban_long["Country Code"].isin(exclude_codes)
]

electricity_long = electricity_long[
    ~electricity_long["Country Code"].isin(exclude_codes)
]
```

## Restructuring, Merging, and Final Data Cleaning

In this phase of the study, the wide-format World Bank indicator datasets (`renewable_long`, `gdp_long`, `urban_long`, and `electricity_long`) must be combined with the carbon emissions data from Our World in Data (`co2`). 

Because all World Bank variables retain an identical `Country Name` column, a standard horizontal merge will generate conflicting duplicate column suffixes (e.g., `Country Name_x`, `Country Name_y`), triggering computational execution errors in pandas. To circumvent this, duplicate descriptive labels are safely dropped immediately prior to merging, using the strict structural index key `['Country Code', 'Year']` as our internal tracking anchor. Furthermore, missing value records are completely dropped post-merge to ensure model equilibrium, preventing computation failures during the matrix inversions in our regression libraries.

```python
co2_cleaned = co2.copy()
```

```python
if 'Code' in co2_cleaned.columns and 'Country Code' not in co2_cleaned.columns:
    co2_cleaned = co2_cleaned.rename(columns={'Code': 'Country Code'})
```

```python
co2_cleaned['Year'] = co2_cleaned['Year'].astype(int)
```

```python
co2_col = [col for col in co2_cleaned.columns if col not in ['Entity', 'Country Code', 'Year', 'Country Name']][0]
co2_cleaned = co2_cleaned.rename(columns={co2_col: 'CO2_Emissions'})
```

```python
data = renewable_long.copy()
data = data.merge(gdp_long.drop(columns=["Country Name"]), on=["Country Code", "Year"], how="inner")
data = data.merge(urban_long.drop(columns=["Country Name"]), on=["Country Code", "Year"], how="inner")
data = data.merge(electricity_long.drop(columns=["Country Name"]), on=["Country Code", "Year"], how="inner")
data = data.merge(co2_cleaned[['Country Code', 'Year', 'CO2_Emissions']], on=["Country Code", "Year"], how="inner")
```

```python
initial_shape = data.shape
data = data.dropna()
final_shape = data.shape

print("==============================================================================")
print("                          DATA MERGING & PURGING REPORT                       ")
print("==============================================================================")
print(f"Total historical entries synthesized:              {initial_shape[0]}")
print(f"Complete country-year rows retained for modeling:  {final_shape[0]}")
print(f"Incomplete/null structural anomalies expunged:     {initial_shape[0] - final_shape[0]}")
print("==============================================================================")
data.head()
```

## Exploratory Data Analysis: Checking Statistical Regression Assumptions

Before constructing statistical models, we must validate key Gauss-Markov assumptions for Ordinary Least Squares (OLS) estimation. 

1. **Multicollinearity Check:** High correlation among independent variables builds unstable model weights, inflating coefficient variances. We compute the Variance Inflation Factor (VIF) for each metric. A standard institutional threshold of $\text{VIF} > 5.0$ signals significant multicollinearity, requiring structural corrections.
2. **Outlier Detection:** Global socioeconomic datasets frequently contain extreme structural leverage profiles (e.g., severe geographic variations in wealth or industrial carbon footprints) that skew classic least-squares equations.

```python
X_cols = ['GDP_per_Capita', 'Urban_Population', 'Electricity_Access', 'CO2_Emissions']
Y_col = 'Renewable_Energy'

X = data[X_cols]
Y = data[Y_col]
```

```python
vif_df = pd.DataFrame()
vif_df["Explanatory Variable"] = X_with_const.columns
vif_df["VIF Vector Value"] = [variance_inflation_factor(X_with_const.values, i) for i in range(X_with_const.shape[1])]

print("==============================================================================")
print("                      MULTICOLLINEARITY DIAGNOSTICS (VIF)                     ")
print("==============================================================================")
print(vif_df.to_string(index=False))
print("==============================================================================")
```

Since all of my actual explanatory variables have VIF scores well under 2.0, I can confidently state that there is no multicollinearity problem in my data. Each variable brings its own unique, independent information to the model, meaning my upcoming regression weights will be stable and reliable!

## Inferential Statistical Modeling: OLS vs. Robust Regression Analysis

Now that I have verified my data does not suffer from multicollinearity, I will run two different regression models to identify the key drivers of Renewable Energy Consumption:

1. **Ordinary Least Squares (OLS) Linear Model:** This sets my baseline. It calculates how my independent variables affect renewable energy by minimizing the total squared errors.
2. **Robust Linear Model (RLM):** Because global data has extreme outliers (like countries with massive wealth or heavy carbon emissions), OLS weights can get pulled out of place. The Robust Regression model uses Huber's M-estimation to automatically down-weight extreme country-level outliers, giving me a more reliable, stable picture of global patterns.

```python
# Framework A: Run the Ordinary Least Squares (OLS) Baseline Model
ols_model = sm.OLS(Y, X_with_const).fit()
print("==============================================================================")
print("                                OLS REGRESSION RESULTS                        ")
print("==============================================================================")
print(ols_model.summary())
```

## Interpretation of the OLS Regression Outputs

I examined the overall performance metrics and the individual coefficients from my OLS baseline model. Here is my analysis of the results:

## Model Fit and Explanatory Power
* **R-squared (0.647):** This means my model explains **64.7%** of the variation in renewable energy consumption shares across countries and years. For a global dataset with over 4,300 rows, this is a strong, healthy fit.
* **F-statistic (2010):** The overall model is highly statistically significant, meaning that these four variables combined are reliable predictors of renewable energy patterns.

## Analysis of Individual Drivers (Coefficients & P-Values)
Every single predictor variable I tested returned a $p$-value of **0.000**, which means all our findings are statistically highly significant.

* **CO₂ Emissions per Capita ($\beta = -1.0131$):** This is the strongest predictor. For every 1-unit increase in metric tons of CO₂ emissions per capita, a country's renewable energy share drops by **1.01%**. This shows a direct trade-off: countries heavily dependent on oil, coal, and gas remain locked into non-renewable energy.
* **Electricity Access ($\beta = -0.6860$):** For every 1% increase in grid access across a population, the share of renewable energy drops by **0.69%**. This indicates that as countries rapidly expanded power grids to meet public demands over the last two decades, they primarily extended traditional fossil fuel infrastructure rather than clean alternatives.
* **Urban Population ($\beta = -0.0715$):** A 1% increase in the urbanization rate is associated with a **0.07%** reduction in renewable energy share, likely because growing, highly concentrated urban centers create intense power demands that renewable grids couldn't instantly supply.
* **GDP per Capita ($\beta = 0.0002$):** While technically positive and significant, the coefficient value is virtually zero ($2.0 \times 10^{-4}$). This signals that raw financial wealth alone doesn't instantly dictate clean energy adoption; instead, infrastructure choices matter much more than just a country's income level.

```python
# Framework B: Run the Robust Linear Model (RLM) using Huber's M-Estimation Weights
rlm_model = sm.RLM(Y, X_with_const, M=sm.robust.norms.HuberT()).fit()
print("==============================================================================")
print("                            ROBUST REGRESSION (RLM) RESULTS                  ")
print("==============================================================================")
print(rlm_model.summary())

# Export the final cleaned analytical dataset as a CSV to feed our web dashboard
data.to_csv("cleaned_sdg7_data.csv", index=False)
print("\n[SUCCESS] Unified dataset saved as 'cleaned_sdg7_data.csv' for the dashboard to read!")
```

Funding must be restricted to deploying decentralized clean energy systems—such as solar mini-grids and wind micro-generation systems—ensuring that newly connected populations bypass fossil fuels entirely.
* **Mandate Decentralized Urban Energy Systems:** To counter the negative impact of urbanization ($\beta = -0.1249$), municipal planning departments must transition from centralized fossil power to localized, smart city configurations. Building codes should mandate industrial-scale rooftop solar configurations, grid-integrated building batteries, and urban waste-to-energy technologies to offset intense, localized urban demand.
* **Establish Carbon Re-investment Mandates:** Because high CO₂ emissions indicate heavy fossil-fuel infrastructure lock-in, market incentives alone are insufficient. Governments must implement carbon pricing mechanisms and cap-and-trade systems, with the tax revenues legally restricted to fund domestic green technology subsidies, clean utility research, and infrastructure transition funds.
* **Prioritize Strategic Structural Policies over Raw GDP Targets:** Given that GDP per capita has a negligible direct effect on clean energy adoption, policymakers must reject the assumption that economic growth will naturally fix environmental problems. Financial strategies should focus on targeted green financing instruments, such as public-private partnerships (PPPs) for grid-scale renewable installations and sovereign green bonds, making clean energy development independent of standard economic growth.

## 11. Study Limitations and Future Research Directions

### Research Limitations
While this study yields robust, statistically significant insights into global energy dynamics, certain structural limitations must be acknowledged:
* **Omission of Regulatory Indicators:** This quantitative framework relies heavily on macroeconomic, infrastructural, and environmental indicators. It does not account for qualitative institutional features, such as the presence of national renewable energy mandates, feed-in tariffs, or clean energy subsidy structures, which heavily influence country-level transitions.
* **Data Incomplete Visual Profiles:** Due to data reporting gaps in the official World Bank and Our World in Data repositories, certain low-income developing nations or conflict zones were dropped during the sequential inner-join merging process, creating a slight bias toward stable reporting nations.
* **Aggregate Resource Treatment:** The dependent variable bundles all renewable sources (solar, wind, biomass, hydro) into one unified percentage metric. This masks sub-sector variations, preventing the model from distinguishing between countries transitioning via modern technologies (solar/wind) versus traditional biomass sources.

### Future Research Directions
To build upon the econometric insights established in this project, future analytical research should explore the following dimensions:
* **Integration of Policy Indication Matrices:** Future iterations could incorporate dummy variables or string-encoded policy indexes (e.g., carbon tax presence, Paris Agreement commitment timelines) to isolate how legislative frameworks interact with economic variables.
* **Time-Lagged Predictive Frameworks:** Implementing lag-variables ($t-5$ or $t-10$ years) for independent variables like GDP per Capita and Urbanization would reveal long-term delayed effects, capturing how long it takes for infrastructural investment to materialize as visible green grid results.
* **Sub-Component Panel Inquiries:** Breaking down the target variable into individual sectors (e.g., tracking Solar Energy consumption separately from Hydroelectric power) would clarify which precise clean technologies respond best to rapid national growth and urbanization.
